## Step 1: Import Required Libraries

First, we need to import the libraries we'll use for data manipulation and visualization.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for better-looking plots
sns.set_style("whitegrid")

print("Libraries imported successfully!")
```

## Step 2: Create a Sample Dataset

A simple dataset about student scores in different subjects.

```python
# Create a dictionary with our data
data = {
    'Subject': ['Mathematics', 'Physics', 'Chemistry', 'Biology', 'English'],
    'Average_Score': [85, 78, 82, 90, 88]
}

# Convert dictionary to a DataFrame
df = pd.DataFrame(data)

print("Dataset created successfully!")
```

## Step 3: Display the DataFrame

Let's look at our data in a tabular format.

```python
# Display the DataFrame
print("Sample Dataset - Student Average Scores by Subject:")
print("="*50)
display(df)

# Show basic information about the dataset
print("\nDataset Information:")
print(f"Number of rows: {len(df)}")
print(f"Number of columns: {len(df.columns)}")
print(f"Column names: {list(df.columns)}")
```

## Step 4: Create a Simple Bar Chart (Using Matplotlib)

Now, let's create our first bar chart!

```python
# Create a figure and axis
plt.figure(figsize=(10, 6))

# Create a simple bar chart
plt.bar(df['Subject'], df['Average_Score'])

# Display the chart
plt.show()
```

## Step 5: Add Labels and Title

Let's make our chart more informative by adding labels and a title.

```python
# Create a figure and axis
plt.figure(figsize=(10, 6))

# Create bar chart
plt.bar(df['Subject'], df['Average_Score'])

# Add labels and title
plt.xlabel('Subject', fontsize=12, fontweight='bold')
plt.ylabel('Average Score', fontsize=12, fontweight='bold')
plt.title('Student Average Scores by Subject', fontsize=14, fontweight='bold')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Display the chart
plt.show()
```

## Step 6: Customize Colors

Let's make our chart more visually appealing with custom colors!

```python
# Create a figure and axis
plt.figure(figsize=(10, 6))

# Define custom colors for each bar
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7']

# Create bar chart with custom colors
bars = plt.bar(df['Subject'], df['Average_Score'], color=colors, edgecolor='black', linewidth=1.5)

# Add labels and title
plt.xlabel('Subject', fontsize=12, fontweight='bold')
plt.ylabel('Average Score', fontsize=12, fontweight='bold')
plt.title('Student Average Scores by Subject', fontsize=14, fontweight='bold', pad=20)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height}',
             ha='center', va='bottom', fontsize=10, fontweight='bold')

# Rotate x-axis labels
plt.xticks(rotation=45, ha='right')

# Add grid for better readability
plt.grid(axis='y', alpha=0.3, linestyle='--')

# Set y-axis limits
plt.ylim(0, 100)

# Adjust layout
plt.tight_layout()

# Display the chart
plt.show()
```

## Step 7: Create a Horizontal Bar Chart

Sometimes horizontal bars are easier to read, especially with long labels.

```python
# Create a figure and axis
plt.figure(figsize=(10, 6))

# Create horizontal bar chart
bars = plt.barh(df['Subject'], df['Average_Score'], color='#3498DB', edgecolor='black', linewidth=1.5)

# Add labels and title
plt.xlabel('Average Score', fontsize=12, fontweight='bold')
plt.ylabel('Subject', fontsize=12, fontweight='bold')
plt.title('Student Average Scores by Subject (Horizontal)', fontsize=14, fontweight='bold')

# Add value labels at the end of each bar
for i, bar in enumerate(bars):
    width = bar.get_width()
    plt.text(width + 1, bar.get_y() + bar.get_height()/2.,
             f'{width}',
             ha='left', va='center', fontsize=10, fontweight='bold')

# Add grid
plt.grid(axis='x', alpha=0.3, linestyle='--')

# Set x-axis limits
plt.xlim(0, 100)

# Adjust layout
plt.tight_layout()

# Display the chart
plt.show()
```

## Step 8: Using Seaborn for Bar Charts

Seaborn provides a higher-level interface that often requires less code.

```python
# Create a figure and axis
plt.figure(figsize=(10, 6))

# Create bar chart using Seaborn
sns.barplot(data=df, x='Subject', y='Average_Score', palette='viridis', edgecolor='black', linewidth=1.5)

# Add labels and title
plt.xlabel('Subject', fontsize=12, fontweight='bold')
plt.ylabel('Average Score', fontsize=12, fontweight='bold')
plt.title('Student Average Scores by Subject (Seaborn)', fontsize=14, fontweight='bold')

# Rotate x-axis labels
plt.xticks(rotation=45, ha='right')

# Set y-axis limits
plt.ylim(0, 100)

# Adjust layout
plt.tight_layout()

# Display the chart
plt.show()
```

### Percentage Stacked Bar Chart

Sometimes it's useful to show proportions rather than absolute values.

```python
# Create horizontal stacked bar chart
plt.figure(figsize=(12, 6))

# Get subjects
subjects = students_df['Subject']
y = np.arange(len(subjects))

# Create horizontal stacked bars
bar1 = plt.barh(y, students_df['Alice'], label='Alice', color='#FF6B6B', edgecolor='black', linewidth=1.2)
bar2 = plt.barh(y, students_df['Bob'], left=students_df['Alice'], label='Bob', color='#4ECDC4', edgecolor='black', linewidth=1.2)
bar3 = plt.barh(y, students_df['Charlie'], left=students_df['Alice'] + students_df['Bob'], label='Charlie', color='#45B7D1', edgecolor='black', linewidth=1.2)

# Add labels and title
plt.xlabel('Total Score', fontsize=12, fontweight='bold')
plt.ylabel('Subject', fontsize=12, fontweight='bold')
plt.title('Student Scores by Subject (Horizontal Stacked)', fontsize=14, fontweight='bold')

# Set y-axis ticks
plt.yticks(y, subjects)

# Add legend
plt.legend(loc='lower right', fontsize=10, framealpha=0.9)

# Add grid
plt.grid(axis='x', alpha=0.3, linestyle='--')

# Adjust layout
plt.tight_layout()

# Display the chart
plt.show()
```

### Stacked Bar Chart - Horizontal Version

Horizontal stacked bars can be easier to read, especially with many categories.

```python
# Create stacked bar chart using Matplotlib
plt.figure(figsize=(12, 6))

# Get subjects
subjects = students_df['Subject']
x = np.arange(len(subjects))

# Create stacked bars
bar1 = plt.bar(x, students_df['Alice'], label='Alice', color='#FF6B6B', edgecolor='black', linewidth=1.2)
bar2 = plt.bar(x, students_df['Bob'], bottom=students_df['Alice'], label='Bob', color='#4ECDC4', edgecolor='black', linewidth=1.2)
bar3 = plt.bar(x, students_df['Charlie'], bottom=students_df['Alice'] + students_df['Bob'], label='Charlie', color='#45B7D1', edgecolor='black', linewidth=1.2)

# Add labels and title
plt.xlabel('Subject', fontsize=12, fontweight='bold')
plt.ylabel('Total Score', fontsize=12, fontweight='bold')
plt.title('Student Scores by Subject (Stacked Bar Chart)', fontsize=14, fontweight='bold', pad=20)

# Set x-axis ticks
plt.xticks(x, subjects, rotation=45, ha='right')

# Add legend
plt.legend(loc='upper left', fontsize=10, framealpha=0.9)

# Add grid
plt.grid(axis='y', alpha=0.3, linestyle='--')

# Adjust layout
plt.tight_layout()

# Display the chart
plt.show()
```

## Step 10: Stacked Bar Charts

Stacked bar charts show the total while breaking down the composition. Perfect for showing how individual contributions add up!

```python
# Reshape data for Seaborn (convert to long format)
students_long = students_df.melt(id_vars='Subject', var_name='Student', value_name='Score')

print("Data in Long Format:")
display(students_long.head(10))

# Create grouped bar chart using Seaborn
plt.figure(figsize=(12, 6))

sns.barplot(data=students_long, x='Subject', y='Score', hue='Student', 
            palette=['#FF6B6B', '#4ECDC4', '#45B7D1'], 
            edgecolor='black', linewidth=1.2)

# Add labels and title
plt.xlabel('Subject', fontsize=12, fontweight='bold')
plt.ylabel('Score', fontsize=12, fontweight='bold')
plt.title('Student Scores by Subject (Seaborn Grouped Chart)', fontsize=14, fontweight='bold')

# Rotate x-axis labels
plt.xticks(rotation=45, ha='right')

# Set y-axis limits
plt.ylim(0, 100)

# Adjust legend
plt.legend(title='Student', fontsize=10, title_fontsize=11, framealpha=0.9)

# Adjust layout
plt.tight_layout()

# Display the chart
plt.show()
```

### Grouped Bar Chart with Seaborn

Seaborn makes grouped bar charts even easier with the `hue` parameter.

```python
# Create grouped bar chart using Matplotlib
import numpy as np

# Set up the figure
plt.figure(figsize=(12, 6))

# Set the width of bars and positions
bar_width = 0.25
x = np.arange(len(students_df['Subject']))

# Create bars for each student
bars1 = plt.bar(x - bar_width, students_df['Alice'], bar_width, label='Alice', color='#FF6B6B', edgecolor='black', linewidth=1.2)
bars2 = plt.bar(x, students_df['Bob'], bar_width, label='Bob', color='#4ECDC4', edgecolor='black', linewidth=1.2)
bars3 = plt.bar(x + bar_width, students_df['Charlie'], bar_width, label='Charlie', color='#45B7D1', edgecolor='black', linewidth=1.2)

# Add labels and title
plt.xlabel('Subject', fontsize=12, fontweight='bold')
plt.ylabel('Score', fontsize=12, fontweight='bold')
plt.title('Student Scores by Subject (Grouped Bar Chart)', fontsize=14, fontweight='bold', pad=20)

# Set x-axis ticks
plt.xticks(x, students_df['Subject'], rotation=45, ha='right')

# Add legend
plt.legend(loc='upper left', fontsize=10, framealpha=0.9)

# Add grid
plt.grid(axis='y', alpha=0.3, linestyle='--')

# Set y-axis limits
plt.ylim(0, 100)

# Adjust layout
plt.tight_layout()

# Display the chart
plt.show()
```
