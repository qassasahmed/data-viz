Below we load the `tips` dataset from Seaborn and produce several common static plots: scatterplot, boxplot, and a heatmap of correlations.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
plt.figure(figsize=(8,6))
plt.title('Total Bill vs Tip (scatter)')
sns.scatterplot(data=df, x='total_bill', y='tip', hue='day', palette='deep')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend(title='Day')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.title('Total Bill Distribution by Day (boxplot)')
sns.boxplot(data=df, x='day', y='total_bill', palette='pastel')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.tight_layout()
plt.show()
```

```python
numerical_cols = ['total_bill', 'tip', 'size']
corr = df[numerical_cols].corr()
plt.figure(figsize=(6,5))
plt.title('Correlation Matrix (static heatmap)')
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.tight_layout()
plt.show()
```

## Notes
- These plots are static and suitable for export to PNG/PDF for reports.
- If you want higher-resolution output for publication, set `plt.figure(dpi=300)` or use `plt.savefig(...)`.
